```python
def IQR(df,feature):
    q1,q3=np.percentile(df[feature],[25,75])
    outliers=[]
    iqr=q3-q1
    upper_bound=q3 + 1.5*(iqr)
    lower_bound=q1 - 1.5*(iqr)
    for val in df[feature]:
        if val < lower_bound or val >upper_bound:
            outliers.append(val)
    return outliers

for feature in continuous_num_features:
    if df[feature].isnull().sum()==0:
        print(f"Outliers of {feature} : {len(IQR(df,feature))}")
```

```python
import statsmodels.api as sm
def Display_QQPlot(df,features):
    for f in features:
        sm.qqplot(data=df[f],line="s")
        plt.title(f)
        plt.show()

Display_QQPlot(df,continuous_num_features)
```

```python
# Using Random Forest Classifier to impute the NaN values in road and seasonalimpact columns 
impute_data=d1.copy()
# road_nan_data=impute_data.drop(['year','demand'],axis=1)
# road_y_find=road_nan_df[road_nan_df['road'].isnull()].loc[:,'road']
# Splitting the dataset into train and test .The test set will be used to predict the NaN of the given feature
def Split_Func(df,drop_f1):
    x_test=df[df[drop_f1].isnull()].drop([drop_f1],axis=1)
    x_trai=df.dropna(subset=[drop_f1])
    x_train=x_trai.drop([drop_f1],axis=1)
    return x_train,x_test
```

```python
def Func_Road(val):
    if val=='nh':
        return 1
    else:
        return 0
road_y_train=impute_data.dropna(subset=['road']).loc[:,'road'].apply(Func_Road) # Converted sh and nh into 0 and 1 in road
```

```python
road_x_train,road_x_test=Split_Func(impute_data,'road')
road_x_train.shape,road_x_test.shape
```

```python
def Compare_Hist(df,new_data,feature):
    plt.figure(figsize=(10,4))
    
    plt.subplot(1,2,1)
    plt.title(f"{feature} With Outliers")
    sns.histplot(data=df,x=feature,kde=True)
    
    plt.subplot(1,2,2)
    plt.title(f"{feature} Without Outliers")
    sns.histplot(data=new_data,x=feature,kde=True)
    plt.show()
```

```python
# This funcion will replace the outliers with the given percentile value
def Change_Outliers(out_lst,data,feature,ptile):
    t=np.percentile(data[feature],ptile)
    for out in out_lst:
        data[feature]=data[feature].replace(out,t)
```

```python
# Splitting into train and test set
X=d5.drop(['tariff'],axis=1)
Y=d5['tariff']
X.shape,Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Scaling of the features
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train2=scaler.fit_transform(X_train)
X_test2=scaler.transform(X_test)
X_train=pd.DataFrame(X_train2,columns=X_train.columns)
X_test=pd.DataFrame(X_test2,columns=X_test.columns)
```

```python
# Building models
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import r2_score
linear=LinearRegression()
ridge=Ridge()
lasso=Lasso()
```

```python
def Model_Performance(x_train,x_test,y_train,y_test,model,r2_score,data):
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    r2=r2_score(y_test,pred)
    adj_r2 = 1 - (1-r2)*(len(data) - 1) / (len(data) - (data.shape[1] - 1) - 1)
    return r2,adj_r2
```

```python
# Linear Regression
linear_r2,linear_adj_r2=Model_Performance(X_train,X_test,y_train,y_test,linear,r2_score,d5)
print(f"r2 score of Linear Regression :{linear_r2}")
print(f"Adjusted r2 score of Linear Regression :{linear_adj_r2}")
```

```python
# Ridge
ridge_r2,ridge_adj_r2=Model_Performance(X_train,X_test,y_train,y_test,ridge,r2_score,d5)
print(f"r2 score of Ridge Regression :{ridge_r2}")
print(f"Adjusted r2 score of Ridge Regression :{ridge_adj_r2}")
```

```python
# Lasso
Lasso_r2,Lasso_adj_r2=Model_Performance(X_train,X_test,y_train,y_test,lasso,r2_score,d5)
print(f"r2 score of Lasso Regression :{Lasso_r2}")
print(f"Adjuested r2 score of Lasso Regression :{Lasso_adj_r2}")
```
